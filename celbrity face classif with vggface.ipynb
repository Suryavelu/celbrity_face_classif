```
from keras_vggface.vggface import VGGFace
from keras.models import Model
from keras.layers import Dense,Flatten,Activation
```

```
face=VGGFace(include_top=False,model='resnet50',input_shape=(224,224,3))
```

```
face.summary()
```

```
for i in face.layers:
  i.trainable=False
```

```
x=Flatten()(face.output)
x=Dense(5)(x)
op=Activation('softmax')(x)
```

```
model=Model(face.input,op)
```

```
model.summary()
```

```
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```
imgen=ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
```

```
train_data=imgen.flow_from_directory('/content/drive/My Drive/Datasets/cropface',(224,224))
```

```
model.fit_generator(train_data,epochs=50)
```

```
model.save('/content/drive/My Drive/Datasets/cropface/face_model.h5')
```
